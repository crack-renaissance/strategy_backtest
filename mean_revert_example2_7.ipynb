# Johansen Test for cointegrating

```python
import numpy as np 
import pandas as pd 
import yfinance as yf 
import matplotlib.pyplot as plt 
```

```python
ewa, ewc = yf.Ticker('EWA') , yf.Ticker('EWC')
```

```python
ewa_df = ewa.history(start='2006-04-26', end='2012-04-09')
ewc_df = ewc.history(start='2006-04-26', end='2012-04-09')
```

```python
plt.plot(ewa_df.index, ewa_df['Open'])
plt.plot(ewc_df.index, ewc_df['Open'])
```
